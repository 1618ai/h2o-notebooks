```python
STOP_WORDS = ["ax","i","you","edu","s","t","m","subject","can","lines","re","what",
               "there","all","we","one","the","a","an","of","or","in","for","by","on",
               "but","is","in","a","not","with","as","was","if","they","are","this","and","it","have",
               "from","at","my","be","by","not","that","to","from","com","org","like","likes","so"]
```

```python
#This is a function that extracts words, while removing words with less than 2 characters, numbers and stop words
def tokenize(sentences, stop_word = STOP_WORDS):
    tokenized = sentences.tokenize("\\W+")
    tokenized_lower = tokenized.tolower()
    tokenized_filtered = tokenized_lower[(tokenized_lower.nchar() >= 2) | (tokenized_lower.isna()),:]
    tokenized_words = tokenized_filtered[tokenized_filtered.grep("[0-9]",invert=True,output_logical=True),:]
    tokenized_words = tokenized_words[(tokenized_words.isna()) | (~ tokenized_words.isin(STOP_WORDS)),:]
    return tokenized_words
```

```python
tokenized = tokenize(loan_df['emp_title'])
```

```python
#tokenized.shape
tokenized.describe()
```

```python
#Word2vec Model
from h2o.estimators.word2vec import H2OWord2vecEstimator
w2vModel = H2OWord2vecEstimator(sent_sample_rate = 0.0, epochs = 15,vec_size = 500)
w2vModel.train(training_frame=tokenized)
#h2o.save_model(model=w2vModel, path="Model")
```

```python
w2vModel.find_synonyms("executive")
```

```python
vec = w2vModel.transform(tokenized, aggregate_method = "AVERAGE")
```

```python
vec.shape
```

```python
#removing words which were not encoded by W2V model
valid_titles = ~ vec["C1"].isna()
```

```python
data = loan_df[valid_titles,["emp_title","annual_inc"]].cbind(vec[valid_titles,:])
```

```python
data_split = data.split_frame(ratios=[0.75], seed=12345)
```

```python
from h2o.estimators import H2OGradientBoostingEstimator

gbm_model = H2OGradientBoostingEstimator()
gbm_model.train(x = vec.names, #+ ["emp_title"],
                y = "annual_inc",
               training_frame = data_split[0],
               validation_frame = data_split[1],
               model_id="w2vModel")
```
